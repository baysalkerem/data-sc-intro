**Turkish Lemmatization**

```python
tokenized_tur_idioms=[]
lemmatized_tur_idioms_list=[]

for i in tur_idioms:
  tokenized_tur_idioms.append(i.split())

for each in tokenized_tur_idioms:
  try:
      for parse in analyzer.analyze(each[-1]):
        part_of_speech = parse[0][4]
        part_of_speech1 = part_of_speech.split(':')
        lemma_word = part_of_speech1[1].split()
        lemma_word1 = lemma_word[1]
        part_of_speech2 = part_of_speech1[2].split("|")
        pos = part_of_speech2[0]
        if pos == "Verb":
          each[-1] = lemma_word1
        if pos != "Verb":
          each[-1] = each[-1]
        
        lemmatized_tur_idioms_list.append(each)       
  except:
    pass

lemmatized_tur_idioms_list
```

```python
lemmatized_tur_idioms = [' '.join(x) for x in lemmatized_tur_idioms_list]

lemmatized_tur_idioms
```
